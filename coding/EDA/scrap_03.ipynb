```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/barry/CodeAcademy/Ideal_dataset/combined_houses/6_months_data.csv')
```

```python
df.time = pd.to_datetime(df.time)
df.index= df.time
```

```python
df= df.drop(columns=['electric-combined-next-hour'], axis=1)
```

```python
resampled_df = df.resample('3H')
resampled_df = resampled_df.agg({
    'temperature_2m (°C)': 'mean',
    'relative_humidity_2m (%)': 'mean',
    'weather_code (wmo code)': 'first',
    'wind_speed_10m (km/h)': 'mean',
    'wind_direction_10m (°)': 'mean',
    'day': 'first',
    'hour': 'first',
    'electric-combined': 'sum',
    'electric-combined-yesterday': 'sum',
    'electric-combined-last-week': 'sum',

})
```

```python
resampled_df['electric-combined_3H-forecast']= resampled_df['electric-combined'].shift(-1)
```
