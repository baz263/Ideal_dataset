```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/barry/CodeAcademy/Ideal_dataset/combined_houses/6_months_data.csv')
```

```python
df.time = pd.to_datetime(df.time)
df.index = df.time
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(data = df, x=df.index.day, y = 'electric-combined', ax=ax)
plt.title('Electricity consumption by day of the month')
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(data = df, x=df.index.day_of_week, y = 'electric-combined', ax=ax)
plt.title('Electricity consumption by day of the week')
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(data = df, x=df.index.month, y = 'electric-combined', ax=ax, order = [9,10,11,12,1,2])
plt.title('Electricity consumption by day of the week')
ax.set_xticklabels(['September', 'October', 'November', 'December', 'January', 'February'])

fig.show()
```
